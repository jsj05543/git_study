```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
x = np.array([1,2,3,4])
y= np.array([2,6,3,2])
ax.plot(x,y,label="x1",linewidth=1,linestyle="dashdot",color="#FF8800",marker="D")

ax.legend()

plt.show()
```

```python
x = [1,2,3]
y1 = [1,2,3]
y2 =[2,3,4]
fig, ax = plt.subplots()
labels=["A","B","C"]
width=0.4
ax.bar(x,y1,tick_label=labels,label="x1",width=width,color="red",align="center",edgecolor="#66ff77")
x2 = [num + width for num in x]
ax.bar(x2,y2,tick_label=labels,label="x2",width=width,bottom=0)
ax.legend(loc='best')
plt.show()
```

```python
x = [1,2,3]
y1 = [1,2,3]
y2 =[2,3,4]
fig, ax = plt.subplots()
labels=["A","B","C"]
y_total = [num1 + num2 for num1,num2 in zip(y1,y2)]
ax.bar(x,y_total,tick_label=labels,label="x1",color="red",align="center")
ax.bar(x,y2,tick_label=labels,label="x2")
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
# ランダムに５０個の要素を生成
np.random.seed(123)
x = np.random.rand(100)
y = np.random.rand(100)

# 散歩図を描画

ax.scatter(x,y,marker="x",label="x1",s=600,c="pink",alpha=0.5, linewidths="2",)
ax.scatter(x,y,marker="D",label="x2",  edgecolors="red",cmap='Blues')
ax.set_title("散布図")
ax.set_xlabel("横")
ax.set_ylabel("縦")
ax.legend()

plt.show()
```

```python
fig, ax =plt.subplots()
# データを生成
np.random.seed(123)
# 平均値
mu = 10
# 標準偏差
sigma =15
x0 = np.random.normal(mu,20,1000)
x1 = np.random.normal(mu,15,1000)
labels=['x0','x1']
n, bins, patches =ax.hist((x0,x1),bins=12,orientation = 'horizontal',stacked=True,label=labels)
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
x0 = np.random.normal(1,2,3)
x1 = np.random.normal(1,3,9)
labels=['A','B']
ax.boxplot((x0,x1),labels = labels,vert =False)
plt.show()
```

```python
labels = ['spam','ham','egg']
x = [10,3,1]
fig, ax = plt.subplots()
explode =[0,0.1,0.2]
#円グラフを描画
ax.pie(x,labels=labels,startangle=90,counterclock=False,shadow=True,autopct='%1.2f%%',explode=explode)

plt.show()
```

```python
fig, axes = plt.subplots(1,6)
axes[0].set_xlabel('折れ線')
axes[1].set_xlabel('棒グラフ')
axes[2].set_xlabel('散布図')
axes[3].set_xlabel('ヒストグラム')
axes[4].set_xlabel('箱ビげ図')
axes[5].set_xlabel('円グラフ')


# グラフを表示
x=[1,1,2]
y=[3,1,4]
axes[0].plot(x,y)
axes[1].bar(x,y)
axes[2].scatter(x,y)
# 平均値
mu = 100
# 標準偏差
sigma =15
x1 = np.random.normal(mu,sigma,1000)
axes[3].hist(x1,10)

axes[4].boxplot(x1)
axes[5].pie(x)
plt.show()
```
