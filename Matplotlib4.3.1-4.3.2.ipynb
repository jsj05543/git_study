# Matplotlib

## 4.3.1 Matplotlibの概要

### Matplotlibとは

- matplotlibはPythonで主に２次元のグラフを描画するためのライブラリ

```python
# matplotlibを利用する
import matplotlib.pyplot as plt
```

```python
import matplotlib.style
# 利用するグラフのスタイルを指定　ggplot
matplotlib.style.use('ggplot')
```

- MATLABスタイル

```python
# データ用意
x = [1,2,3]
y = [1,4,9]
# 折れ線グラフを描画
plt.plot(x,y)
# グラフにタイトルを設定
plt.title=('MATLAB-style')
# グラフを表示
plt.show()
```

- オブジェクト指向スタイル

```python
# データ用意
x = [1,2,3]
y = [1,4,9]
# 描画オブジェクト(fig)とサブプロント(ax)を作成
fig, ax = plt.subplots()
# 折れ線グラフを描画
ax.plot(x,y)
# グラフにタイトルを設定
ax.set_title('00P-style')
# グラフを表示
plt.show()
```

## 4.3.2 描画オブジェクト

```python
# ２つのサブプロットを配置
fig, ax = plt.subplots(2)
# グラフを表示
plt.show()
```

```python
# ２x２のサブプロットを配置
fig, ax = plt.subplots(2,2)
# グラフを表示
plt.show()
```

```python
# 1x２のサブプロットを配置(nrows,ncols)を指定して描画
fig, ax = plt.subplots(nrows =1,ncols=2)
# グラフを表示
plt.show()
```

```python
# グラフスタイル一覧を表示
print(matplotlib.style.available)
```

```python
# グラフのスタイルをclassicに指定
matplotlib.style.use('seaborn-deep')
# 描画オブジェクト、サブプロット
fig, ax =plt.subplots()
#データ用意
ax.plot([1,2])
plt.show()
```

```python
fig, axes = plt.subplots(nrows =1,ncols=2)
# サブプロットにタイトルを設定
axes[0].set_title('subplot title 0')
axes[1].set_title('subplot title 1')
# X軸、Y軸ラベル設定
axes[0].set_xlabel('x label 0')
axes[0].set_ylabel('y label 0')
axes[1].set_xlabel('x label 1')
axes[1].set_ylabel('y label 1')
# 描画オブジェクトにタイトルを設定
fig.suptitle('figure title')
plt.show()
```

```python
fig, ax = plt.subplots(nrows =1,ncols=1)
# 汎例用のラベルを設定
ax.plot([1,2,3],[2,4,9],label="legend label")
#ラベル表示位置
ax.legend(loc='best')
plt.show()
```

```python
fig, ax = plt.subplots(nrows =1,ncols=1)
# 汎例用のラベルを設定
ax.plot([1,2,3],[2,4,9],label="legend label")
#ラベル表示位置
ax.legend(loc='lower right')
plt.show()
```
