## 4.3.3 グラフの種類と出力方法

### `各種グラフの描画方法`

| No. | グラフ名 | 関数 |　引数 |  説明 |
| ----- | ------- | ------- | ----- | ------- | 
| 1| 折れ線グラフ | matplotlib.pyplot.plot | x,y,label | x軸位置、y軸位置、ラベル|
| 2| 棒グラフ | matplotlib.pyplot.bar | x,y,tick_label,label | x軸位置、y軸位置、外ラベル、内ラベル|
| 3| 横向き棒グラフ | matplotlib.pyplot.barh | x,y,label | x軸位置、y軸位置、ラベル|
| ４| 散歩図 | matplotlib.pyplot.scatter | x,y,label,marker | x軸位置、y軸位置,ラベル、表記|
| 5| ヒストグラム | matplotlib.pyplot.hist | m,sigma,x,bins,orientation,label,stacked | m:平均、sigma:標準偏差,x:乱数,bins:ビンの数<br>横向,レベル,true:積み上げ|
| 6| 箱びげ図 | matplotlib.pyplot.boxplot | x,labels,vert | x軸位置、ラベル、false横向き|
| 7| 折円グラフ | matplotlib.pyplot.pie | x,<br>labels,<br>startangle,<br>counterclock,<br>shadow,<br>autopct,<br>explode | データ、<br>ラベル、<br>開始位置、<br>falseは時計回り、<br>trueは影あり、<br>パーセント表示、<br>切り出し|

### 1.折れ線グラフ
* matplotlib.pyplot.plot

```python
fig, ax =plt.subplots()
x = [1,2,3]
y1 = [1,2,3]
y2 = [3,1,2]
ax.plot(x,y1)
ax.plot(x,y2)

plt.show()
```

### 2.折れ線グラフの値間隔を細かくした曲線グラフ

```python
import numpy as np
x = np.arange(0.0,15.0,0.1)
y1 = np.sin(x)
y2 = np.cos(x)
fig, ax =plt.subplots()
ax.plot(x,y1,label='sin')
ax.plot(x,y2,label='cos')

ax.legend()

plt.show()
```

### 3.棒グラフ
* matplotlib.pyplot.bar

```python
fig, ax =plt.subplots()
x = [1,2,3]
y = [10,2,3]
labels = ['spam','ham','egg']
# ラベルを指定
ax.bar(x,y,tick_label =labels)

plt.show()
```

```python
# 横向き棒グラフ
fig, ax =plt.subplots()
x = [1,2,3]
y = [10,2,3]
labels = ['spam','ham','egg']
# ラベルを指定
ax.barh(x,y,tick_label =labels)

plt.show()
```

```python
# 複数のグラフを並べて描画
fig, ax =plt.subplots()
x = [1,2,3]
y1 =[10,2,3]
y2 = [5,3,6]
labels= ['spam','ham','egg']
# 棒グラフの幅を指定して描画
width =0.4
# 幅を指定して描画
ax.bar(x,y1,width = width,tick_label = labels,label = 'y1')
# 幅分ずらして棒グラフを描画
x2 = [num + width for num in x]
ax.bar(x2,y2,width=width, label = 'y2')

ax.legend()

plt.show()
```

```python
# 積み上げ棒グラフ
fig, ax =plt.subplots()
x = [1,2,3]
y1 = [10,2,3]
y2 = [5,2,3]
labels =['spam','ham','egg']
# y1とy2を足した値を格納
y_total = [num1 + num2 for num1,num2 in zip(y1,y2)]

# y1とy2を足した高さの棒グラフを描画
ax.bar(x,y_total,tick_label = labels, label = 'y1')
ax.bar(x,y2,label = 'y2')

ax.legend()

plt.show()
```

### 4.散歩図
* matplotlib.pyplot.scatter

```python
fig, ax = plt.subplots()

# ランダムに５０個の要素を生成
np.random.seed(123)
x = np.random.rand(50)
y = np.random.rand(50)

# 散歩図を描画
ax.scatter(x,y)

plt.show()
```

```python
fig, ax = plt.subplots()

# ランダムに５０個の要素を生成
np.random.seed(123)
x = np.random.rand(50)
y = np.random.rand(50)

# 散歩図を描画「マーカー別」
# 下向き三角
ax.scatter(x[0:10],y[0:10],  marker='v',label = 'triang down')
# 仰向き三角
ax.scatter(x[10:20],y[10:20],marker='^',label = 'triangle up')
 # 正方形
ax.scatter(x[20:30],y[20:30],marker='s',label = 'square')
# 星形
ax.scatter(x[30:40],y[30:40],marker='*',label = 'star')
# X
ax.scatter(x[40:50],y[40:50],marker='x',label = 'x')

ax.legend()

plt.show()
```

### 5.ヒストグラム
* matplotlib.pyplot.hist

```python
# データを生成
np.random.seed(123)
# 平均値
mu = 100
# 標準偏差
sigma =15
x = np.random.normal(mu,sigma,1000)

fig, ax = plt.subplots()

# ヒストグラムを描画
n, bins, patches = ax.hist(x)

plt.show()
```

```python
for i, num in enumerate(n):
    print('{:.2f} - {:.2f}:{}'.format(bins[i],bins[i + 1],num))
```

```python
fig, ax = plt.subplots()

# ビンの数を指定して描画
n, bins, patches =ax.hist(x,bins=25)

plt.show()
```

```python
fig, ax = plt.subplots()

# 横向きのヒストグラムを描画
ax.hist(x, orientation = 'horizontal')

plt.show()
```

```python
# データを生成
np.random.seed(123)
# 平均値
mu = 100

# 異なる標準偏差でデータを生成
x0 = np.random.normal(mu,20,1000)
x1 = np.random.normal(mu,15,1000)
x2 = np.random.normal(mu,10,1000)

fig, ax = plt.subplots()
labels = ['x0','x1','x2']

# ３つのデータのヒストグラムを描画
ax.hist((x0,x1,x2), label = labels)
ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots()
labels = ['x0','x1','x2']
#積み上げたヒストグラムを描画
ax.hist((x0,x1,x2),label = labels, stacked =True)
ax.legend()

plt.show()
```

### 6.箱びげ図
* matplotlib.pyplot.boxplot

```python
# 異なる標準偏差でデータを生成
np.random.seed(123)
x0 = np.random.normal(0,10,500)
x1 = np.random.normal(0,15,500)
x2 = np.random.normal(0,20,500)

fig, ax = plt.subplots()
labels =['x0','x1','x2']

#箱びげ図を描画
ax.boxplot((x0,x1,x2),labels=labels)

plt.show()
```

```python
fig, ax = plt.subplots()
labels =['x0','x1','x2']

#横向きの箱びげ図を描画
ax.boxplot((x0,x1,x2),labels=labels,vert =False)

plt.show()
```

### 7.円グラフ 
* matplotlib.pyplot.pie

```python
labels = ['spam','ham','egg']
x = [10,3,1]
fig, ax = plt.subplots()

#円グラフを描画
ax.pie(x,labels=labels)

plt.show()
```

```python
fig, ax = plt.subplots()

#円グラフを描画
ax.pie(x,labels=labels)
# アスペクト比を保持して描画
ax.axis('equal')

plt.show()
```

```python
fig, ax = plt.subplots()

#円グラフを描画、開始位置を12時から
ax.pie(x,labels=labels, startangle=90)
# アスペクト比を保持して描画
ax.axis('equal')

plt.show()
```

```python
fig, ax = plt.subplots()

#円グラフを描画、開始位置を１２時から、時計回り
ax.pie(x,labels=labels, startangle=90, counterclock =False)
# アスペクト比を保持して描画
ax.axis('equal')

plt.show()
```

```python
fig, ax = plt.subplots()

#円グラフを描画、開始位置を１２時から、時計回り,影をつけ、数字をグラフ上表示
ax.pie(x,labels = labels, startangle=90, counterclock =False, shadow=True,autopct='%1.2f%%')
# アスペクト比を保持して描画
ax.axis('equal')

plt.show()
```

```python
# 1番目の要素を切り出す
explode =[0,0.1,0.2]

fig, ax = plt.subplots()

#円グラフを描画、開始位置を１２時から、時計回り,影をつけ、数字をグラフ上表示、要素を切り出す
ax.pie(x,labels = labels, startangle=90, counterclock =False,shadow=True, autopct='%1.2f%%',explode=explode)
# アスペクト比を保持して描画
ax.axis('equal')

plt.show()
```

### 複数のグラフを組み合わせる

```python
# 棒グラフと折れ線グラフを組合せる
fig, ax = plt.subplots()
# データ作成
x1 = [1,2,3]
y1 = [5,2,3]
x2 = [1,2,3,4]
y2 =[8,5,4,6]

# 棒グラフを描画
ax.bar(x1,y1,label ='y1')
# 折れ線グラフを描画
ax.plot(x2,y2,label ='y2')
ax.legend()

plt.show()
```

```python
# 近似曲線、ヒストグラムを組合せる
np.random.seed(123)

#正規乱数を生成
x = np.random.randn(1000)

fig, ax = plt.subplots()

# ヒストグラムを描画
counts, edges, patches = ax.hist(x, bins = 25)

# 近似曲線に持ちしている点を求める
x_fit =(edges[:-1] + edges[1:]) / 2
# 近似曲線をプロット
y = 1000 * np.diff(edges) * np.exp(-x_fit**2 / 2) /np.sqrt(2 * np.pi)

ax.plot(x_fit,y)

plt.show()
```
