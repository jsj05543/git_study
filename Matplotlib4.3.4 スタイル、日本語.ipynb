## 4.3.4 スタイル

```python
# matplotlibを利用する
import matplotlib.pyplot as plt
import matplotlib.style
# 利用するグラフのスタイルを指定　ggplot
matplotlib.style.use('ggplot')
```

### 色の設定

```python
fig ,ax =plt.subplots()

# 線の色を名前で指定
ax.plot([1,3],[3,1],label ='日本', color='aqua')

#１６新数のRGBで指定
ax.plot([1,3],[1,3],label='#000FF',color='#0000FF')

#vRGBAをfloatで指定
ax.plot([1,3],[2,2],label = '(0.1,0.2,0.5,0.3)',color=(0.1, 0.2, 0.5, 0.5))

# 凡例を表示
ax.legend()

plt.show()
```

```python
fig, ax =plt.subplots()

# 塗りつぶりの色指定
ax.bar([1],[3], color='aqua')

# 塗りつぶりの色を枠線の色指定
ax.bar([2],[4], color='aqua',edgecolor='black')

plt.show()
```

### 線のスタイル

```python
fig, ax =plt.subplots()

# 5.5ポイントの幅の線で描画
ax.plot([1,3],[3,1], linewidth=5.5,label ='5.5')
# 10ポイントの幅の線で描画
ax.plot([1,3],[1,3], linewidth=10,label ='10')

ax.legend()

plt.show()
```

```python
fig, ax =plt.subplots()

# 破線で描画
ax.plot([1,3],[3,1], linestyle='--',label ='dashed')
# 一点鎖線で描画
ax.plot([1,3],[1,3], linestyle='-.',label ='dashdot')
# 点線で描画
ax.plot([1,3],[2,2], linestyle=':',label ='dotted')

ax.legend()

plt.show()
```

### フォント

```python
fig, ax = plt.subplots()

# フォントのスタイルを指定
ax.set_xlabel('xlabel',family='fantasy',size=20,weight='bold');
ax.set_ylabel('ylabel',family='cursive',size=40,weight='light');
ax.set_title('graph title',family='monospace',size=25,weight='heavy');

#plt.show()
```

```python
#css
fontdict={
    'family':'fantasy',
    'size':20,
    'weight':'normal',
}

fig, ax = plt.subplots()

# 辞書形式でフォントのスタイルを指定
ax.set_xlabel('xlabel',fontdict=fontdict)
ax.set_ylabel('ylabel',fontdict=fontdict)
#個別指定したsizeで上書き可能
ax.set_title('graph title',fontdict=fontdict,size=40)

plt.show()
```

### テキスト描画

```python
fig, ax = plt.subplots()

# Textというテキストを描画
ax.text(0.2,0.8,'Sample Text',size=40)
plt.show()
```

### pandasのオブジェクトからグラフ描画


```python
import pandas as  pd
import matplotlib.style
import matplotlib.pyplot as plt

#スタイル設定
matplotlib.style.use('ggplot')


# DataFrameを作成
df = pd.DataFrame({'A':[1,2,3],'B':[3,1,2]})
# 折れ線グラフを描画
df.plot()
plt.show()
```

```python
import numpy as np

# ランダムに３行２列のデータを生成してDataFrameを生成
np.random.seed(123)
df2 = pd.DataFrame(np.random.rand(3,2), columns=['　日本語','y2'])

#棒グラフを描画
df2.plot.bar()
plt.show()
```

```python
#積み上げ棒グラフを描画
df2.plot.bar(stacked =True)
plt.show()
```

```python
#横向きグラフを描画
df2.plot.barh()
plt.show()
```
